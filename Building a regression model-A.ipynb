<h1>Building a Baseline Model</h1>

**Importing the necessary libraries**

```python
import keras 
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.models import Sequential 
from keras.layers import Dense
```

**Importing the data**

```python
df = pd.read_csv("https://cocl.us/concrete_data")
df.head()
```

```python
df_columns = df.columns
predictors = df[df_columns[df_columns != 'Strength']]
target = df[["Strength"]]
```

**Building the model**

```python
input_shape = predictors.shape[1] #number of predictors 
def baseline_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(input_shape,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

**Normalizing the Data**

```python
predictors_normalized = predictors
```

```python
target_normalized = target 
```

**Training the model**

```python
mse_list = []

# Repeat process 50 times
for i in range(50):
    # Splitting data into training and test sets
    predictors_train, predictors_test, target_train, target_test = train_test_split(
    predictors_normalized, target_normalized, test_size=0.3, random_state=42)
    
    # Creating the model
    model = baseline_model()
    # Training the model on the training data
    model.fit(predictors_normalized, target_normalized, validation_split=0.3, epochs=50, verbose=1)    
    
    # Evaluating the model on the test data and compute MSE
    target_prediction = model.predict(predictors_normalized)
    mse = mean_squared_error(target_normalized, target_prediction)
    
    # Appending MSE to the list
    mse_list.append(mse)
```

***Calculating the mean of the MSE***

```python
avg_mse = np.mean(mse_list)
print(f'The average mse is {avg_mse}')
```

***Calculating the std of the MSE***

```python
std_mse = np.std(mse_list)
print(f'The standard deviation of the mse is {std_mse}')
```
